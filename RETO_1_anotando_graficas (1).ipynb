```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
```

```python
#montar drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/Remoto diabetes-clean.csv', index_col=0)

df.head()
```

### a)

```python
fig=plt.figure(figsize=(10,5))
glucosa=sns.distplot(df['glucose'], kde=False, norm_hist=False)
glucosa.set_title('Distribución de glucosa')
glucosa.set_xlabel('Glucosa')
glucosa.set_ylabel('Frecuencia')

```

### b)

```python
pregnancy_counts = df['pregnancies'].value_counts()
```

```python
fig=plt.figure(figsize=(10,5))
pregn=sns.barplot(x=pregnancy_counts.index,y= pregnancy_counts)
pregn.set_title('Distribución de embarazos')
pregn.set_xlabel('Embarazos')
pregn.set_ylabel('Frecuencia')
```

### c)

```python
healthy = df[df['outcome'] == 0]
sick = df[df['outcome'] == 1]
```

```python
fig=plt.figure(figsize=(10,5))
sanos=sns.distplot(healthy['glucose'], hist=False, kde_kws={'shade': True})
senfermos=sns.distplot(sick['glucose'], hist=False, kde_kws={'shade': True})
sanos.set_title('Distribución de glucosa en personas sanas')
sanos.set_xlabel('Glucosa')
sanos.set_ylabel('Frecuencia')
senfermos.set_title('Distribución de glucosa en personas enfermas')
#poner leyenda de enfermos y sanos
senfermos.legend(['Enfermos', 'Sanos'])
```

### d)

```python
fig=plt.figure(figsize=(10,5))
sns.distplot(sick['blood_pressure'], hist=False, kde_kws={'shade': True})
sns.distplot(sick['glucose'], hist=False, kde_kws={'shade': True})
sns.distplot(sick['insulin'], hist=False, kde_kws={'shade': True})
#poner leyenda de las variables
plt.legend(['Presión arterial', 'Glucosa', 'Insulina'])
```
