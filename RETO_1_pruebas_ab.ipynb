```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#leer dataset
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/Remoto ab_test-sales_pipeline-2.csv')
df.head()
```

```python
conversion=df.groupby('test')['conversion'].mean()
conversion
```

```python
#efectividad y proporcion
if conversion['A']>conversion['B']:
  proporcion=(conversion['A']/conversion['B'])-1
  print('flujo A es mas efectivo')
elif conversion['A']<conversion['B']:
  proporcion=(conversion['B']/conversion['A'])-1
  print('flujo B es mas efectivo', proporcion)
else:
  print('los dos flujos son iguales')
```

```python
from itertools import permutations
# nuevos grupos
R=10000
permutations=[]
combinado=df['test'].values
# Get the size of group A based on the 'conversion' column
sizeA=df[df['test']=='A'] ['conversion'].shape[0]

for i in range(R):
  np.random.shuffle(combinado)
  nuevoA=combinado[:sizeA]
  nuevoB=combinado[sizeA:]
  # Convert 'A' and 'B' to numerical values before calculating the mean
  # For example, you can assign 1 to 'A' and 0 to 'B'
  numeric_nuevoA = [1 if x == 'A' else 0 for x in nuevoA]
  numeric_nuevoB = [1 if x == 'A' else 0 for x in nuevoB]

  resultado=np.mean(numeric_nuevoA)-np.mean(numeric_nuevoB)
  permutations.append((resultado))

grupos=pd.DataFrame({'A':pd.Series(dtype='float64'),'B':pd.Series(dtype='float64')})
grupos.head()
```
