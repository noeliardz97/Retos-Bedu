```python
import pandas as pd
#leer datos de wine /content/drive/MyDrive/Colab Notebooks/Datasets/winequality-red.csv
wine = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/winequality-red.csv', sep=';')
wine.head()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#establecer estilo
sns.set(style="darkgrid")
plt.figure(figsize=(10, 6))
#scaterplot
# Check if 'alcohol' is the correct column name, it might be 'alcohol' or something similar
# If it is different, replace 'alcohol' with the actual column name below
# The 'data' parameter should be set to the DataFrame you want to plot.
# The 'x' and 'y' parameters should specify the column names for the x and y axes.
scat=sns.scatterplot(data=wine, x='alcohol', y='pH', hue='quality',palette='viridis') # Changed x to 'alcohol' assuming that is the intended column
scat.set_title('Relación PH,Calidad y acidez')
scat.set_xlabel('Alcohol' )
scat.set_ylabel('Densidad')
plt.show()
```

```python
#scatter plots por categoria de calidad
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#establecer estilo
sns.set(style="darkgrid")
plt.figure(figsize=(10, 6))
#scaterplot
calidad= wine['quality'].unique()
calidad.sort()

for quality in calidad:
  plt.figure(figsize=(10, 6))
  scat=sns.scatterplot(data=wine[wine['quality'] == quality], x='alcohol', y='pH', hue='quality',palette='viridis') # Changed x to 'alcohol' assuming that is the intended column
  scat.set_title(f'Calidad {quality}')
  scat.set_xlabel('Alcohol' )
  scat.set_ylabel('Densidad')
  plt.show()
```
