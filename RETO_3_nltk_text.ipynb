```python
#importar archivo
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/Remoto amazon_fine_food_reviews-clean.csv')
df.head()
```

```python
from nltk import word_tokenize,Text
from nltk.corpus import stopwords
import nltk
nltk.download('punkt')
```

```python
#metodo word_tokenize
df['tokenized_text'] = df['text'].apply(word_tokenize)
df.head()
```

```python
all_words = [word for sublist in df['tokenized_text'] for word in sublist]
```

```python
text=Text(all_words)
```

```python
#CONCORDANCIA BOY
text.concordance('boy',lines=10)

```

```python
#CONCORDANCIA GIRL
text.concordance('girl',lines=10)
```

```python
#Calcular la riqueza
len(set(all_words))/len(all_words)
```
