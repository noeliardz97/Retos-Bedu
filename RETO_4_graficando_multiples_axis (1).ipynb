En el Reto pasado generamos tablas de contingencia para segmentar nuestros datos usando dos o más variables categóricas. El problema con nuestras gráficas es que resultaba muy incómodo e ineficiente compararlas. Una manera mucho más efectiva de comparar varias gráficas es graficarlas en la misma figura. Tu Reto consiste en lo siguiente:

1. Lee el dataset 'melbourne_housing-clean.csv'.
2. Crea una tabla de contingencia donde el índice esté dado por la variable 'region_name' y las columnas por la variable 'type'. Esto nos va a dar una tabla donde tenemos la frecuencia de cada tipo de propiedad segmentada por regiones.
3. Genera una figura con 8 subplots para hacer gráficas de barras de cada región en nuestro dataset. Cada gráfica tendrá 3 barras, correspondiendo a los tipos 'h', 't' y 'u'.
4. Asegúrate de que compartan eje `x` y `y` para que sea fácil compararlas. Además agrega anotaciones y títulos donde creas necesario.
5. Describe tus hallazgos y comparte con tus compañeros qué podrían significar.

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/melbourne_housing-clean.csv', index_col=0)
regionname_type = pd.crosstab(df['region_name'], [df['type']])

fig, axes = plt.subplots(2, 4, figsize=(15, 8), sharex=True, sharey=True)

# Use the 'x' and 'y' keyword arguments to specify the data columns
sns.barplot(x=regionname_type.loc['Eastern Metropolitan'].index, y=regionname_type.loc['Eastern Metropolitan'], ax=axes[0, 0])
sns.barplot(x=regionname_type.loc['Eastern Victoria'].index, y=regionname_type.loc['Eastern Victoria'], ax=axes[0, 1])
sns.barplot(x=regionname_type.loc['Northern Metropolitan'].index, y=regionname_type.loc['Northern Metropolitan'], ax=axes[0, 2])
sns.barplot(x=regionname_type.loc['Northern Victoria'].index, y=regionname_type.loc['Northern Victoria'], ax=axes[0, 3])
sns.barplot(x=regionname_type.loc['South-Eastern Metropolitan'].index, y=regionname_type.loc['South-Eastern Metropolitan'], ax=axes[1, 0])
sns.barplot(x=regionname_type.loc['Southern Metropolitan'].index, y=regionname_type.loc['Southern Metropolitan'], ax=axes[1, 1])
sns.barplot(x=regionname_type.loc['Western Metropolitan'].index, y=regionname_type.loc['Western Metropolitan'], ax=axes[1, 2])
sns.barplot(x=regionname_type.loc['Western Victoria'].index, y=regionname_type.loc['Western Victoria'], ax=axes[1, 3])

axes[0, 0].set_title('Eastern Metropolitan')
axes[0, 1].set_title('Eastern Victoria')
axes[0, 2].set_title('Northern Metropolitan')
axes[0, 3].set_title('Northern Victoria')
axes[1, 0].set_title('South-Eastern Metropolitan')
axes[1, 1].set_title('Southern Metropolitan')
axes[1, 2].set_title('Western Metropolitan')
axes[1, 3].set_title('Western Victoria')

plt.tight_layout()
plt.show()
````
