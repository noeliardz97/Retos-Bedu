```python
#import dataset
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/winequality-red.csv', delimiter=';')
df.head()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn.metrics import make_scorer, mean_squared_error # Import make_scorer and mean_squared_error

#validacion cruzada de k iteraciones
x= df[['density', 'citric acid']]
y= df['fixed acidity']

modelo=   LinearRegression()
kf= KFold(n_splits=5, shuffle=True, random_state=42)

scores= cross_val_score(modelo, x, y, cv=kf, scoring='neg_mean_squared_error')
mse_score=make_scorer(mean_squared_error, greater_is_better=False) # Now make_scorer is defined
scores= cross_val_score(modelo, x, y, cv=kf, scoring=mse_score)

mean=np.mean(scores)
std=np.std(scores)
#imprimir
print("Mean Squared Error:", mean)
print("Standard Deviation:", std)

# Fit the model to the entire dataset before calculating R2 and MSE
modelo.fit(x, y)  # This line is added to fit the model

#imprimir r2
r2=modelo.score(x,y)
print("R2:", r2)
#imprimir mse
mse=mean_squared_error(y,modelo.predict(x))
print("MSE:", mse)
```
